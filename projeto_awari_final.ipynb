```python
#biblioteca warnings limpa
#biblioteca statsmodels que faz previsão em estatistica

import warnings
import statsmodels.api as sm
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def teste (dataset):
  dados_filtrados_teste = adfuller(dataset,autolag = 'AIC')
  print('1. ADF :',dados_filtrados_teste[0])
  print('2. P.Value :',dados_filtrados_teste[1])
  print('3. Num of Lags :',dados_filtrados_teste[2])
  print('4. Num of observations used for ADF regression and cristical values calculation :',dados_filtrados_teste[3])
  print('5. Cristical values:')
  for l, val in dados_filtrados_teste[4].items():
    print("\t",l, ": ", val)
```

```python
teste(dados_filtrados['temperatura'])
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
print(dados_filtrados.shape)
train=dados_filtrados.iloc[:-30]
teste=dados_filtrados.iloc[-30:]
print(train.shape, teste.shape)
```

```python
model=ARIMA(dados_filtrados['temperatura'],order=(0,1,0))
model=model.fit()
model.summary()
```

```python
start=len(dados_filtrados)
end=len(dados_filtrados)+len(teste)-1
pred=model.predict(start=start,end=end,typ='levels')
print(pred)
```

```python
pred.plot(legend=True)
teste['temperatura'].plot(legend=True)
```
